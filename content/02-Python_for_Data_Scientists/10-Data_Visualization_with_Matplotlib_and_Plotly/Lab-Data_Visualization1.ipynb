## Visualizing Titanic Dataset

Load the titanic dataset, use seaborn to draw violin plots to visualize the following:

1. Survived vs PClass - Split the plot(hue) by gender.
2. Survived Vs SibSp (Number of siblings spouse) - hue by gender 

The dataset is available at:
"https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv"




```python
import pandas as pd

train_data = pd.read_csv("https://raw.githubusercontent.com/agconti/kaggle-titanic/master/data/train.csv")
train_data.head()
```

```python
%matplotlib inline
import seaborn as sns

sns.violinplot(y='Survived', x='Pclass', hue='Sex', split=True, data=train_data)
```

```python
sns.violinplot(y='Survived', x='SibSp', hue='Sex', data=train_data)
```
