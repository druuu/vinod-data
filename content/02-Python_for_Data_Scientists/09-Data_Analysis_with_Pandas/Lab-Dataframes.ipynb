## Simplify the Grad Students Dataset

Given a dataset about Graduate Students and their college majors and other statistics about the number of graduates, number employed etc, write python code to extract the following infromation from the dataset and write to a output file - "grad_simplified.csv". Use pandas library (Dataframe) for this purpose.

The criteria for output extraction is:

1. Only include the columns: Major_Code, Major, Major_Category, Grad_employed and Grad_unemployed
2. Only include the Majors that have either 'ENGINEERING', 'TECHNOLOGIES' or 'SCIENCE' (case insensitive) in its major  field.
3. Exclude any Major_category that has 'Social Science' or 'Arts' in the name of Major_category.




```python
# Import the library
import pandas as pd

grad_url = "https://raw.githubusercontent.com/colaberry/538data/master/college-majors/grad-students.csv"
grad_students = pd.read_csv(grad_url)

#Write your code here or in other code cells down
grad_students.head()
```

```python
columns_reduced = grad_students[['Major_code', 'Major', 'Major_category', 'Grad_employed', 'Grad_unemployed']]
print(columns_reduced.shape)
columns_reduced.head()
```

```python
rows_reduced1 = columns_reduced[columns_reduced.Major.str.contains('ENGINEERING') | columns_reduced.Major.str.contains('TECHNOLOGIES') | columns_reduced.Major.str.contains('SCIENCE')]
print(rows_reduced1.shape)
rows_reduced1.head()
```

```python
rows_reduced2 = rows_reduced1[rows_reduced1.Major_category.str.contains('Social Science') | rows_reduced1.Major_category.str.contains('Arts')].index
rows_reduced2
```

```python
rows_reduced3 = rows_reduced1.drop(rows_reduced2)
rows_reduced3.head()
```

```python
rows_reduced3.shape
```

```python
rows_reduced3.to_csv("grad_simplified.csv",index=False)
```

## Simplify the NYSE dataset for Facebook and Netflix

```python
import numpy as np
import pandas as pd
import csv
```

```python
url1 = "https://raw.githubusercontent.com/manirv/dsin100daysclass/master/nyse_split_adjusted.csv"
nyse = pd.read_csv(url1)

nyse.head()
```

```python
nyse_nflx = nyse[nyse.symbol.str.contains('NFLX')]
nyse_nflx.head()
```

```python
nflx_close = nyse_nflx[['close']]
nflx_close.index = nyse_nflx['date']
nflx_close = nflx_close.rename(columns={'close':'NFLX'})
nflx_close.head()
```

```python
nyse_fb = nyse[nyse.symbol.str.contains('FB')]
nyse_fb.head()
```

```python
fb_close = nyse_fb[['close']]
fb_close.index = nyse_fb['date']
fb_close = fb_close.rename(columns={'close':'FB'})
fb_close.head()
```

```python
url2="https://raw.githubusercontent.com/Vinod-Kumar-G/dsin100daysclass/master/SPY.csv"
spy = pd.read_csv(url2)
spy.head()
```

```python
adj_close = spy[['Adj Close']]
adj_close.index = spy['Date']
adj_close = adj_close.rename(columns={'Adj Close':'SPY'})
adj_close.head()
```

```python
dates = spy.iloc[:,0:1]
dates.head()
```

```python
stock = pd.DataFrame(index=dates['Date'])
stock.head()
```

```python
stock = stock.join(adj_close)
stock = stock.join(fb_close)
stock = stock.join(nflx_close)
stock.head()
```
